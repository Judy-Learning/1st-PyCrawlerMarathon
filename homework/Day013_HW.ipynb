### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

```python
newest = soup.find(class_='r-ent')
print(newest.find(class_='author').text.strip())
print(newest.find(class_='title').text.strip())
print(newest.find(class_='date').text.strip())
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
## Sample
for d in soup.find_all(class_='title'):
    print(d.text.replace('\t', '').replace('\n', ''))
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
        print('作者: ' + r.find(class_='article-meta-value').text)
    except:
        continue
        
```

```python
# 他人
arts = soup.find_all(class_='r-ent')
for art in arts:
    print(art.find(class_='author').text.strip())
    print(art.find(class_='title').text.strip())
    print(art.find(class_='date').text.strip())
```

```python
infos = []
for d in soup.find_all(class_="title"):
    try:
        r = BeautifulSoup(s.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
        author, tag, title, date = [info.text for info in r.find_all(class_="article-meta-value")]
        #print(date, author, title)
        infos.append([date, author, title])
        #break
    except:
        continue
```

```python
import pandas as pd
columns = ["Date", "Author", "Title"]
pd.DataFrame(infos, columns=columns)
```
