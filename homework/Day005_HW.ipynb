```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
### 如何判斷熱門/非熱門文章？ Ans:用網址的popular=true 或是　false
```

```python
#熱門文章網址
a = "https://www.dcard.tw/_api/forums/pet/posts?popular=true"
import requests
r1 = requests.get(a)
response_1 = r1.text

import json
data_1 = json.loads(response_1)

#我的→會有Error
# for d1 in data_1:
#     a = print(d1['commentCount'])
# type(a)
# a = list(a) TypeError: 'NoneType' object is not iterable

#他人
num_data_1 = len(data_1)
commentCount_1 = 0
likeCount_1 = 0

#迭代去累加
for d in data_1:
    commentCount_1 += d["commentCount"]
    likeCount_1 += d["likeCount"]
      
#=====================================================    
#非熱門文章網址
b = "https://www.dcard.tw/_api/forums/pet/posts?popular=false"
import requests
r2 = requests.get(b)
response_2 = r2.text

import json
data_2 = json.loads(response_2)

num_data_2 = len(data_2)
commentCount_2 = 0
likeCount_2 = 0

for d in data_2:
    commentCount_2 += d["commentCount"]
    likeCount_2 += d["likeCount"]
```

```python
print("熱門")
print("平均留言人數:", commentCount_1 / num_data_1)
print("平均按讚人數:", likeCount_1 / num_data_1)

print("\n非熱門")
print("平均留言人數:", commentCount_2/ num_data_2)
print("平均按讚人數:", likeCount_2 / num_data_2)
```
